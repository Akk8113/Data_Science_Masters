Q1. Scatter Plot for Age and Fare Columns in the Titanic Dataset
python
Copy code
import plotly.express as px
import seaborn as sns

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Create the scatter plot
fig = px.scatter(titanic, x='age', y='fare', title='Scatter Plot of Age vs Fare')
fig.show()

Q3. Histogram for "Sex" vs "Total Bill" with Additional Parameters
python
Copy code
import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Create the histogram
fig = px.histogram(tips, x='sex', y='total_bill', color='day', pattern_shape='smoker', title='Histogram of Total Bill by Sex and Day')
fig.show()

Q5. Distplot
In Plotly Express, the functionality similar to distplot is provided by histogram with a density option. Plotly Express does not have a direct distplot function, but you can use histogram to create a similar plot.

python
Copy code
import plotly.express as px

# Load the Titanic dataset
titanic = px.data.titanic()

# Create the histogram (distplot equivalent)
fig = px.histogram(titanic, x='age', marginal='density', title='Distribution of Age with Density Plot')
fig.show()
