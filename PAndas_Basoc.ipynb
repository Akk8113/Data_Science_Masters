from googleapiclient.discovery import build
import csv

# Set up the YouTube Data API
api_key = 'YOUR_API_KEY'  # Replace with your API key
youtube = build('youtube', 'v3', developerKey=api_key)

# Parameters for the search
search_query = "YOUR_SEARCH_QUERY"  # Replace with your search query
max_results = 5

# Search for videos
request = youtube.search().list(
    part="snippet",
    q=search_query,
    maxResults=max_results,
    type="video"
)
response = request.execute()

# Extract data
videos = []
for item in response['items']:
    video_id = item['id']['videoId']
    title = item['snippet']['title']
    thumbnail_url = item['snippet']['thumbnails']['high']['url']
    published_at = item['snippet']['publishedAt']
    video_url = f"https://www.youtube.com/watch?v={video_id}"
    
    # Fetch video details to get the view count
    video_details = youtube.videos().list(part='statistics', id=video_id).execute()
    view_count = video_details['items'][0]['statistics'].get('viewCount', '0')
    
    videos.append([video_url, title, thumbnail_url, view_count, published_at])

# Write to a CSV file
with open('youtube_videos.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Video URL", "Title", "Thumbnail URL", "View Count", "Published At"])
    writer.writerows(videos)

print("Data has been written to youtube_videos.csv")
Notes:
Replace YOUR_API_KEY with your actual API key from the Google Cloud Console.
Replace YOUR_SEARCH_QUERY with your desired search term.
The script will save the data into a CSV file named youtube_videos.csv.
Make sure to follow YouTube's API terms of use and respect their guidelines on data usage.
