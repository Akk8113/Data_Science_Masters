Q2: Distribution of flight prices.

python
Copy code

# Plot histogram

plt.ylabel('Frequency')
plt.show()
Explanation: The histogram visualizes how flight prices are distributed across different price ranges.

Q3: Range of prices.

python
Copy code
# Calculate range, minimum, and maximum
price_min = df_flight['price'].min()
price_max = df_flight['price'].max()
price_range = price_max - price_min

print(f"Minimum price: {price_min}")
print(f"Maximum price: {price_max}")
print(f"Range of prices: {price_range}")
Explanation: This calculates the minimum and maximum prices and the range between them.

Q4: Price variation by airline.

python
Copy code

Explanation: The boxplot helps compare the distribution of prices for different airlines.
