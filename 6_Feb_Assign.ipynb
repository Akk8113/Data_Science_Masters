```python
Q1: Function to Flatten a List and Return the Product of All Numbers
To solve this, we need to create a function that handles nested structures and extracts numeric values. Hereâ€™s a function that accomplishes this:

python
Copy code
from functools import reduce
import operator

def flatten_and_product(input_list):
    flat_list = []

    def flatten(element):
        if isinstance(element, (list, tuple, set)):
            for item in element:
                flatten(item)
        elif isinstance(element, dict):
            for key, value in element.items():
                if isinstance(key, (int, float)):
                    flat_list.append(key)
                flatten(value)
        elif isinstance(element, (int, float)):
            flat_list.append(element)

    flatten(input_list)
    product = reduce(operator.mul, flat_list, 1)
    return product

list1 = [1, 2, 3, 4, [44, 55, 66, True], False, (34, 56, 78, 89, 34), {1, 2, 3, 3, 2, 1},
         {1: 34, "key2": [55, 67, 78, 89], 4: (45, 22, 61, 34)}, [56, 'data science'], 'Machine Learning']

product = flatten_and_product(list1)
print("The product of all numbers is:", product)
This function defines a nested flatten function that traverses the list, tuples, sets, and dictionaries to extract numeric values. It then uses the reduce function from the functools module to compute the product of these numbers.

Q2: Encrypt a Message Using a Custom Logic
To encrypt a message where each letter is replaced by its corresponding letter from the end of the alphabet (e.g., 'a' -> 'z', 'b' -> 'y', etc.), and spaces are replaced with dollar signs, we can use the following function:

python
Copy code
def encrypt_message(message):
    encrypted_message = []
    for char in message:
        if char.isalpha():
            # Check if the character is uppercase or lowercase
            if char.islower():
                encrypted_char = chr(219 - ord(char))
            else:
                encrypted_char = chr(155 - ord(char))
        elif char == ' ':
            encrypted_char = '$'
        else:
            encrypted_char = char  # Punctuation marks unchanged
        encrypted_message.append(encrypted_char)
    return ''.join(encrypted_message)

input_sentence = "I want to become a Data Scientist."
encrypted_sentence = encrypt_message(input_sentence)
print("Encrypted Sentence:", encrypted_sentence)
This code works as follows:

For each character in the input string, it checks if the character is an alphabet letter.
If it's a lowercase letter, it calculates the corresponding letter from the end of the alphabet using chr(219 - ord(char)). The number 219 comes from adding 'a' (97) to 'z' (122) and subtracting 1.
If it's an uppercase letter, it uses chr(155 - ord(char)). The number 155 is similarly derived for uppercase letters.
Spaces are replaced with dollar signs, and punctuation marks are left unchanged.
These solutions should meet your requirements for both the list processing and message encryption tasks!
```
