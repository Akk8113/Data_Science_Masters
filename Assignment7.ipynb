Q1)

A database is a collection of data that is organized and stored electronically in a computer system. It is a software application that manages data storage, retrieval, and manipulation. Different types of databases exist, including SQL (Structured Query Language) and NoSQL (Not Only SQL) databases.

SQL databases are organized in tables with rows and columns, and data is stored in a structured format. SQL databases are designed to handle complex queries and transactions, and are commonly used in enterprise applications. Examples of SQL databases include MySQL, Oracle, and Microsoft SQL Server.

NoSQL databases, on the other hand, are designed to handle unstructured or semi-structured data, such as JSON or XML documents. NoSQL databases are often used in web applications, mobile applications, and big data analytics. Examples of NoSQL databases include MongoDB, Cassandra, and Couchbase.

Here are some examples of DDL statements:

* CREATE TABLE table_name (

* ALTER TABLE table_name ADD column_name data_type;

DDL statements are typically executed by a database administrator or other authorized user. They are not typically used by end-users, who typically use a data manipulation language (DML) to interact with the database

DML stands for Data Manipulation Language. It is a language used to interact with data in a database. DML statements are used to retrieve, insert, update, and delete data. DML statements are typically used by end-users to interact with the database.

Here are some examples of DML statements:

* SELECT * FROM table_name;

* INSERT INTO table_name (column_name1, column_name2, ...) VALUES (value1, value2, ...);

* UPDATE table_name SET column_name1 = value1, column_name2 = value2, ... WHERE condition;

* DELETE FROM table_name WHERE condition;

DML statements are typically executed by a database client application, such as a spreadsheet or a reporting tool. The client application sends the DML statement to the database server, which executes the statement and returns the results to the client application.

NSERT, UPDATE, and DELETE are three of the most basic operations that can be performed on a database.

* INSERT adds a new row to a table.

* UPDATE modifies an existing row in a table.

* DELETE removes a row from a table.

* SELECT * FROM table_name;

A primary key is a column or set of columns in a database table that uniquely identifies each row in the table. A foreign key is a column or set of columns in a database table that references the primary key of another table.

Primary keys are used to enforce referential integrity in a database, which means that data in one table cannot be changed or deleted if it is referenced by data in another table. Foreign keys are used to create relationships between tables in a database.

For example, consider a database that has two tables: a `customers` table and an `orders` table. The `customers` table has a primary key column called `customer_id`, and the `orders` table has a foreign key column called `customer_id` that references the `customer_id` column in the `customers` table. This means that each order in the `orders` table must be associated with a customer in the `customers` table.

Primary keys and foreign keys are important concepts in database design, and they can help to ensure that data in a database is accurate and consistent.

The order of execution of SQL clauses in an SQL query is as follows:

1. FROM
2. WHERE
3. GROUP BY
4. HAVING
5. SELECT
6. ORDER BY
7. LIMIT

The FROM clause specifies the tables to be used in the query. The WHERE clause specifies the conditions that must be met for a row to be included in the result set. The GROUP BY clause groups the rows in the result set by one or more columns. The HAVING clause specifies the conditions that must be met for a group to be included in the result set. The SELECT clause specifies the columns to be included in the result set. The ORDER BY clause specifies the order in which the rows in the result set should be returned. The LIMIT clause specifies the maximum number of rows to be returned in the result set.

The following is an example of an SQL query that uses all of the clauses:


        SELECT name, age
        FROM users
        WHERE age > 18
        GROUP BY age
        HAVING COUNT(*) > 1
        ORDER BY age DESC
        LIMIT 10;
​


This query will return the names and ages of the 10 youngest users who have more than one entry in the users table.
