1. Dividing a Number by 3 Until Less Than or Equal to 10
To find how many times a given number can be divided by 3 before it is less than or equal to 10, you can use a while loop. Here's an example:

python
Copy code
number = 100  # You can replace 100 with any number
count = 0

while number > 10:
    number /= 3
    count += 1

print(f"The number can be divided {count} times before it becomes less than or equal to 10.")
In this example, number starts at 100 and is divided by 3 in each iteration of the loop. The loop continues until number is less than or equal to 10. The variable count keeps track of how many times the division occurred.

2. Why and When to Use while Loop in Python
Why Use while Loop:

The while loop is useful when the number of iterations is not known beforehand and depends on a condition being met. It is often used when you need to repeatedly execute a block of code as long as a certain condition remains true.
When to Use while Loop:

Use a while loop when you have a condition that needs to be checked before each iteration, and you don't know in advance how many times you want to loop.
Example:

python
Copy code
counter = 0
limit = 5

while counter < limit:
    print(f"Counter: {counter}")
    counter += 1
In this example, the loop prints the value of counter and increments it by 1 in each iteration. The loop continues until counter is no longer less than limit.

3. Use Nested while Loop to Print 3 Different Patterns
Pattern 1: Right-Angle Triangle

python
Copy code
n = 5
i = 1

while i <= n:
    j = 1
    while j <= i:
        print("*", end=" ")
        j += 1
    print()
    i += 1
Pattern 2: Inverted Right-Angle Triangle

python
Copy code
n = 5
i = n

while i > 0:
    j = 1
    while j <= i:
        print("*", end=" ")
        j += 1
    print()
    i -= 1
Pattern 3: Pyramid

python
Copy code
n = 5
i = 1

while i <= n:
    spaces = n - i
    while spaces > 0:
        print(" ", end=" ")
        spaces -= 1
    stars = 1
    while stars <= (2 * i - 1):
        print("*", end=" ")
        stars += 1
    print()
    i += 1
