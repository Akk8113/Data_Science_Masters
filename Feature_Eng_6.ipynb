import numpy as np

study_hours = np.array([1, 2, 3, 4, 5])
exam_scores = np.array([50, 55, 65, 70, 80])

# Calculate Pearson correlation coefficient
correlation_coefficient = np.corrcoef(study_hours, exam_scores)[0, 1]
print("Pearson Correlation Coefficient:", correlation_coefficient)
Interpretation:
The correlation coefficient ranges from -1 to 1. A coefficient close to 1 indicates a strong positive linear relationship, while a coefficient close to -1 indicates a strong negative linear relationship. A coefficient around 0 suggests no linear relationship.

from scipy.stats import spearmanr

sleep_hours = np.array([6, 7, 5, 8, 4])
job_satisfaction = np.array([8, 7, 6, 9, 5])

# Calculate Spearman's rank correlation coefficient
rho, p_value = spearmanr(sleep_hours, job_satisfaction)
print("Spearman's Rank Correlation Coefficient:", rho)
Interpretation:
Spearman's rank correlation ranges from -1 to 1, similar to Pearson’s but it assesses monotonic relationships, not just linear ones.

Q3: Pearson vs. Spearman's Rank Correlation
Pearson Correlation Coefficient and Spearman's Rank Correlation can be compared to understand the type of relationship between variables.

Example Calculation:

python
Copy code
from scipy.stats import spearmanr

exercise_hours = np.array([1, 2, 3, 4, 5])  # Example data
bmi = np.array([30, 28, 25, 22, 20])         # Example data

# Pearson correlation
pearson_corr = np.corrcoef(exercise_hours, bmi)[0, 1]

# Spearman's rank correlation
spearman_corr, _ = spearmanr(exercise_hours, bmi)

print("Pearson Correlation Coefficient:", pearson_corr)
print("Spearman's Rank Correlation Coefficient:", spearman_corr)
Comparison:

Pearson measures linear relationships.
Spearman measures monotonic relationships.
