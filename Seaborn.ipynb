Que 1: Five Plots using Seaborn and Their Uses
Scatter Plot (scatterplot):

Use: Visualizes the relationship between two numerical variables. It helps identify trends, correlations, and outliers.
Bar Plot (barplot):

Use: Displays the distribution of categorical data with corresponding numerical values. It is useful for comparing the quantities of different categories.
Box Plot (boxplot):

Use: Summarizes the distribution of a dataset using quartiles, highlighting the median, quartiles, and potential outliers. It is useful for identifying the central tendency and variability of data.
Heatmap (heatmap):

Use: Represents data in a matrix format, with colors indicating different data values. It is useful for showing relationships between two categorical variables and visualizing correlations in a dataset.
Pair Plot (pairplot):

Use: Displays pairwise relationships between multiple variables in a dataset. It is useful for exploring distributions, relationships, and correlations in the data.
Que 2: Line Plot for the "fmri" Dataset
python
Copy code
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri = sns.load_dataset('fmri')

# Plot a line plot using x = "timepoint" and y = "signal"
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title('FMRI Signal Over Time')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.show()
Que 3: Box Plots for the "titanic" Dataset
python
Copy code
# Load the "titanic" dataset
titanic = sns.load_dataset('titanic')

# Box plot for 'age'
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Age Distribution by Pclass')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Fare Distribution by Pclass')

plt.tight_layout()
plt.show()
Que 4: Histogram for the "diamonds" Dataset
python
Copy code
# Load the "diamonds" dataset
diamonds = sns.load_dataset('diamonds')

# Histogram for 'price' with hue for 'cut'
sns.histplot(diamonds, x='price', hue='cut', multiple='stack')
plt.title('Distribution of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()

Que 6: Heatmap for the "flights" Dataset
python
Copy code
# Load the "flights" dataset
flights = sns.load_dataset('flights')

# Pivot the data to a matrix form
flights_pivot = flights.pivot('month', 'year', 'passengers')

# Plot a heatmap
sns.heatmap(flights_pivot, cmap='coolwarm', annot=True, fmt='d')
plt.title('Heatmap of Monthly Flight Passengers')
plt.show()
