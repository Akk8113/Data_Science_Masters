Q1. What is Matplotlib? Why is it used? Name Five Plots Using Pyplot
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used for data visualization and graphical plotting in the data analysis community. The library's most popular module, Pyplot, offers a collection of functions that make plotting simple and intuitive.

Matplotlib is used for:

Visualizing data to understand trends and patterns.
Creating publication-quality plots.
Customizing plots to fit specific needs.
Five types of plots that can be created using the Pyplot module are:

Line Plot
Scatter Plot
Bar Plot
Histogram
Pie Chart
Q2. What is a Scatter Plot?
A scatter plot is a type of plot that shows the relationship between two variables using dots, where each dot represents an observation. It helps to identify correlations or patterns between the variables.

Here's the code to generate and plot the data:

python
Copy code
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
Q3. Why is the subplot() Function Used? Four Line Plots Using subplot()
The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to organize your plots in a grid, making it easier to compare different plots side-by-side.

Here's the code to create four line plots using subplot():

python
Copy code
import numpy as np
import matplotlib.pyplot as plt

# Data for four line plots
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Plotting
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("Line 4")

plt.tight_layout()
plt.show()
Q4. What is a Bar Plot? Plot a Bar and Horizontal Bar Plot
A bar plot is a type of plot that presents categorical data with rectangular bars, where the length of each bar represents the value of the category. It is used to compare different categories or to show changes over time.

Here's the code for a bar plot and a horizontal bar plot:

python
Copy code
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical Bar Plot
plt.bar(company, profit)
plt.title("Company Profits - Bar Plot")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

# Horizontal Bar Plot
plt.barh(company, profit)
plt.title("Company Profits - Horizontal Bar Plot")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
Q5. What is a Box Plot? Plot a Box Plot
A box plot (also known as a whisker plot) is a graphical representation of the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It helps to identify outliers and understand the data distribution's spread and skewness.

Here's the code to plot a box plot:

python
Copy code
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2], labels=["Box 1", "Box 2"])
plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show()
