Is Python case sensitive when dealing with identifiers?

Yes, Python is case-sensitive when dealing with identifiers. For example, Variable and variable would be considered two different identifiers.
What is the correct extension of the Python file?

The correct extension for a Python file is .py.
Is Python code compiled or interpreted?

Python is primarily an interpreted language, but it is also compiled to bytecode before being interpreted by the Python virtual machine.
