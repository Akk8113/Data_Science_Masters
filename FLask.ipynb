Q1. Explain GET and POST methods.
GET Method:

The GET method is used to request data from a server. It appends the data to the URL in the form of query strings.
It is often used to retrieve data and does not modify the state of the server. For example, fetching a webpage or data from an API.
GET requests can be cached, bookmarked, and remain in the browser history, but they have a limitation on the amount of data they can send because URLs have a length restriction.
POST Method:

The POST method is used to send data to a server to create or update a resource. The data sent to the server with a POST request is included in the body of the request, not the URL.
It is commonly used when submitting form data or uploading files.
POST requests are not cached, do not remain in the browser history, and cannot be bookmarked.
Q2. Why is request used in Flask?
In Flask, request is an object that encapsulates the HTTP request data sent by the client to the server. It includes information such as:

Form data: Data submitted through HTML forms.
Query string parameters: Data sent in the URL.
Headers: Metadata sent by the client, like content type, user agent, etc.
Files: Any files uploaded by the client.
request allows the server to access the data sent by the client, making it possible to process inputs, handle user interactions, and generate appropriate responses.
