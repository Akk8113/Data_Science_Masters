Q2. Difference Between loc and iloc in pandas.DataFrame
loc: Accesses rows and columns by label/index names. It includes the last element in the range.
iloc: Accesses rows and columns by integer-location based indexing. It does not include the last element in the range.
Q3. Reindexing and Accessing Rows
python
Copy code
reindex = [3, 0, 1, 2]
new_df = df.reindex(reindex)

# Output for new_df.loc[2]
print(new_df.loc[2])

# Output for new_df.iloc[2]
print(new_df.iloc[2])
Observation:
new_df.loc[2] uses the label-based indexing and will return the row with the original index label 2.
new_df.iloc[2] uses integer-based indexing and will return the third row in the reindexed DataFrame, which corresponds to the original index 1.

Q6. Windows Function in Pandas
The windows function in pandas is used for performing operations over a sliding window of a specified width. It helps in calculating rolling statistics such as mean, sum, etc.

Types of windows functions include:

rolling(): Provides rolling window calculations.
expanding(): Provides expanding window calculations.
ewm(): Provides exponentially weighted window calculations.

Q8. Difference Between Two Dates
python
Copy code
import pandas as pd

def calculate_date_difference():
    date1 = pd.Timestamp(input("Enter the first date (YYYY-MM-DD): "))
    date2 = pd.Timestamp(input("Enter the second date (YYYY-MM-DD): "))

    difference = date2 - date1
    days = difference.days
    hours = difference.total_seconds() // 3600
    minutes = difference.total_seconds() // 60

    print(f"Difference: {days} days, {int(hours)} hours, and {int(minutes)} minutes")

Q9. Convert a Column to Categorical Data Type
python
Copy code
import pandas as pd

def convert_to_categorical():
    file_path = input("Enter the file path: ")
    column_name = input("Enter the column name to convert: ")
    category_order = input("Enter the category order (comma-separated): ").split(',')

    df = pd.read_csv(file_path)
    df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)
    sorted_df = df.sort_values(by=column_name)
    
    print(sorted_df)

convert_to_categorical()
Q10. Visualize Sales Data Using Stacked Bar Chart
python
Copy code
import pandas as pd
import matplotlib.pyplot as plt

def visualize_sales_data():
    file_path = input("Enter the file path: ")
    df = pd.read_csv(file_path)
    
    df.set_index('Date', inplace=True)
    df.groupby('Product').resample('M').sum().unstack().plot(kind='bar', stacked=True)
    
    plt.title('Sales Data')
    plt.xlabel('Product')
    plt.ylabel('Sales')
    plt.show()

visualize_sales_data()
Q11. Calculate Mean, Median, and Mode from Student Data
python
Copy code
import pandas as pd
from scipy import stats

def calculate_statistics():
    file_path = input("Enter the file path of the CSV file containing the student data: ")
    df = pd.read_csv(file_path)
    
    mean = df['Test Score'].mean()
    median = df['Test Score'].median()
    mode = stats.mode(df['Test Score'])[0][0]
    
    print("+-----------+--------+")
    print("| Statistic | Value  |")
    print("+-----------+--------+")
    print(f"| Mean      | {mean:.1f}   |")
    print(f"| Median    | {median}    |")
    print(f"| Mode      | {mode}    |")
    print("+-----------+--------+")

calculate_statistics()
